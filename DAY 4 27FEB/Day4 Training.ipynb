```python
#Fantasy Battle Arena

class Character():
    def __init__(self,name,health,attack_power,defense,speed):
        self.name = name
        self.health = health
        self.attack_power = attack_power
        self.defense = defense
        self.speed = speed

    def attack(self,target):
        damage=max(target.attack_power - target.defense,1)
        target.take_damage(damage)
        return damage

    def take_damage(self,amount):
        self.health -=amount
        if self.health<0:
            self.health=0

    def is_alive(self):
        return self.health>0

class Warrior(Character):
    def __init__(self, name):
        super().__init__(name,health=150,attack_power=30,defense=20,speed=20)
        self.rage=0
    def taking_damage(self,rage):
        self.rage +=amount
        if self.rage>50:
            self.attack_power *=2
        super().take_damage(amount)
    
    def attack(self,target):
        if(self.health<30):
            self.attack_power *=2
        damage = super().attack(target)
        if self.health<30:
            self.attack_power = 30
        return damage

class Mage(Character):
    def __init__(self, name):
        super().__init__(name,health=100,attack_power=40,defense=10,speed=15)
        self.mana=30
        
    def attack(self,target):
        if (self.mana > 15):
            self.mana -=15
            damage=40
            self.take_damage(5)
        else:
            damage=super().attack(target)
        target.take_damage(damage)
        return damage

class Archer(Character):
    def __init__(self, name):
        super().__init__(name,health=120,attack_power=25,defense=15,speed=25)
        self.attack_count = 0
        
    def attack(self,target):
        self.attack_count+=1
        damage = super().attack(target)

        if self.attack_count%3==0:
            damage *=2
        target.take_damage(damage)
        return damage

def Determine_turn(p1,p2):
    if p1.speed > p2.speed:
        return p1,p2
    elif p1.speed < p2.speed:
        return p2,p1
    else:
        return p1,p2

def start(p1,p2):
    player1,player2 =Determine_turn(p1,p2)

    while player1.is_alive() and player2.is_alive():
        damage = player1.attack(player2)
        if not player2.is_alive():
            print(f"{player2.name} is defeated!")
            break

        player1, player2 = player2,player1
    
    if player1.is_alive():
        winner = player1
        print(f"{winner.name} wins the battle")
    else:
        winner = player2
        print(f"{winner.name} wins the battle")

if __name__=="__main__":
    thor=Warrior("Thor")
    gandalf=Mage("Gandalf")
    alex=Archer("Alex")

    start(alex,thor)   
```
